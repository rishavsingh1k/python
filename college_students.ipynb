```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv("C:/Users/KIIT0001/Desktop/Path/college_student_placement_dataset.csv")
x=dataset.iloc[:,[1,2,3,4,6,7]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans= KMeans(n_clusters=4, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(x)
print(y_kmeans)
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='cluster_1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='pink',label='cluster_2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='magenta',label='cluster_3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='orange',label='cluster_4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('studence performance')
plt.ylabel('Communication skills')
plt.legend()
plt.show()
```
